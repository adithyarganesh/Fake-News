```python
import pandas as pd
import re
from sklearn.feature_extraction.text import  TfidfVectorizer
import nltk
from sklearn import feature_extraction
from nltk.stem import WordNetLemmatizer 
import numpy as np
```

```python
def clean(s):
    return " ".join(re.findall(r'\w+', s, flags=re.UNICODE)).lower()
```

```python
_wnl = nltk.WordNetLemmatizer()
wnl = WordNetLemmatizer()
def normalize_word(w):
    return _wnl.lemmatize(w).lower()


def get_tokenized_lemmas(s):
    return [normalize_word(t) for t in nltk.word_tokenize(s)]
```

```python
def remove_stopwords(l):
    # Removes stopwords from a list of tokens
#     for w in l:
#     if w not in feature_extraction.text.ENGLISH_STOP_WORDS : #or w!='not':
#         return w
    return [w for w in l if w not in feature_extraction.text.ENGLISH_STOP_WORDS or w == 'not']
```

```python
def mega_clean(string):
    string = clean(string)
#     string = normalize_word(string)
    bits = get_tokenized_lemmas(string)
    bits = remove_stopwords(bits)
    
    lem_vect = []
    for word in bits: 
        lem_vect.append(wnl.lemmatize(word, pos='v'))
#     print(lem_vect)
    c =  mapping(lem_vect)
#     print("lalalallalala" + c)
    return c
```

```python
def mapping(lem_vect):
    v = TfidfVectorizer()
    v.fit(lem_vect)
    text = v.vocabulary_
#     print(text)
    sent = []
    for i in text.keys():
        i = clean(i)
        sent.append(i)
        sent.append(' ')
    xx = ["0"]
    xx[0]= ''.join(sent)
#     print("xooxoxoxox" + xx[0])
    return xx[0]
```

```python
########################################################
df = pd.read_csv("fake_or_real_news.csv")
qq = df["text"][2]
arr = re.split('.\n\n',qq)
# print(arr)
sentences = []
for string in arr:
#     print(string)
    x = mega_clean(string)
    sentences.append(x)
#     print("\n\n\nFINALLLL" + x)
# print(arr)
# print("\n\n\n")
# print(sentences) 
```

```python
vv = TfidfVectorizer()
vv.fit(sentences)
mapp = vv.vocabulary_
vv = vv.transform(sentences)
armap = vv.toarray()
```

```python
sums = 0
for i in armap:
    sums  = sums + i
```

```python
minimum = np.min(sums)
```

```python
city = {}
count = 0
for i in sums:
    city[count] = i-minimum
    count = count+1
    
```

```python
total = 0
count = 0
for val in city:
    total = total + city[val]
    count = val

avg = total/count
```

```python
for val in city:
    if city[val] <= avg:
        city[val] = 0
```
